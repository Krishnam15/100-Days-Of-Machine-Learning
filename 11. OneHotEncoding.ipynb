```python
# since brand has so many unique values. We are grouping the uncommon brands to one column.
# setting threshold = 100 for uncommon. Means brands with less than 100 values are considered as uncommon.
# doing above shits coz in One Hot Encoding, the numerical code for each unique value contains the same no. of digits as the no of unique values.
# if we will have so many values, it will lead to a very big encoding. Hence grouping together the uncommon brands.
```

```python
counts = cars['brand'].value_counts()
threshold = 100
repl = counts[counts<=threshold].index
repl #containing brands having value less than 100
```

```python
pd.get_dummies(cars['brand'].replace(repl,'Uncommon')).sample(5)
```
