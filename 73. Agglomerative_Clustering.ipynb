```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\KIIT\Desktop\100 Days of ML\Datasets\customer-spendings.csv")
```

```python
df.head()
```

```python
data = df.iloc[:,3:].values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure()
plt.title("Customer Dendogram")
dendo = shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
labels_ = cluster.fit_predict(data)
```

```python
labels_
```

```python
plt.scatter(data[:,0], data[:,1],c=cluster.labels_,cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=5)
labels = km.fit_predict(data)
```

```python
labels
```

```python
plt.scatter(data[labels==0,0],data[labels==0,1],color='blue')
plt.scatter(data[labels==1,0],data[labels==1,1],color='red')
plt.scatter(data[labels==2,0],data[labels==2,1],color='green')
plt.scatter(data[labels==3,0],data[labels==3,1],color='yellow')
plt.scatter(data[labels==4,0],data[labels==4,1],color='orange')
```

```python
fig, ax = plt.subplots(1,2,figsize = (12,5))

ax[0].scatter(data[:,0], data[:,1],c=cluster.labels_,cmap='rainbow')
ax[0].set_title('Results of Agglomerative Clustering')

ax[1].scatter(data[labels==0,0],data[labels==0,1],color='blue')
ax[1].scatter(data[labels==1,0],data[labels==1,1],color='red')
ax[1].scatter(data[labels==2,0],data[labels==2,1],color='green')
ax[1].scatter(data[labels==3,0],data[labels==3,1],color='yellow')
ax[1].scatter(data[labels==4,0],data[labels==4,1],color='orange')
ax[1].set_title('Results of K-Means Clustering')

plt.tight_layout
plt.show()
```
