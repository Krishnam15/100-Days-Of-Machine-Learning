```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import SimpleImputer, MissingIndicator
```

```python
df = pd.read_csv(r"C:\Users\KIIT\Desktop\100 Days of ML\Datasets\titanic.csv",usecols=['Age','Fare','Survived'])
```

```python
df.sample(5)
```

```python
df.isnull().mean()*100
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf,y_train)

y_pred = clf.predict(X_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
mi = MissingIndicator()
mi.fit(X_train)
```

```python
mi.features_
# the below output shows that both columns 0 (age) and 1 (fare) have missing values.
```

```python
X_train_missing = mi.transform(X_train)
```

```python
X_train_missing
```

```python
X_train[['Age_NA','Fare_NA']] = X_train_missing
```

```python
X_train
```

```python
X_test_missing = mi.transform(X_test)
```

```python
X_test[['Age_NA','Fare_NA']] = X_test_missing
```

```python
X_test
```

```python
si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X_train_trf2,y_train)
```

```python
y_pred = clf.predict(X_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)

# Missing Indicator tends to improve accuracy
```

```python
#default method to call Missing Indicator within the SI directly without calling MI seperately.

si = SimpleImputer(add_indicator=True) 
X_train = si.fit_transform(X_train)
X_test = si.transform(X_test)
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf2,y_train)

y_pred = clf.predict(X_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
