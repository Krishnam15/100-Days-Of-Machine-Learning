```python
titanic = pd.read_csv(r"C:\Users\KIIT\Desktop\100 Days of ML\Datasets\titanic.csv")
```

```python
#bivariate analysis
sns.scatterplot(x=tips['total_bill'],y=tips['tip']) 
# From the plot we can see that it follows a linear relation. Tip increases with an increase in total bill.
```

```python
# multivariate analysis - analyzing 5 columns.
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
# Numerical-Categorical bivariate analysis using barplot
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

```python
# distplot for numerical-categorical bivariate analysis
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False) # hist=False to turn off the histograms.
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)

# The plot shows the probability density function for survival with age.
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Pclass').mean('Survived')['Survived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked').mean('Survived')['Survived']*100)
```
