```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
X=iris.data
```

```python
y=iris.target
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 80,50

# By default, plot_tree() generates a small-sized plot.
# If your decision tree is deep or complex, the default plot will be hard to read, with overlapping nodes and tiny text.
# Setting rcParams['figure.figsize'] = (80, 50) makes the plot much larger, allowing better readability

plot_tree(clf)
```
