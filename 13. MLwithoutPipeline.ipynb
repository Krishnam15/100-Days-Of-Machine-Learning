```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv(r"C:\Users\KIIT\Desktop\100 Days of ML\Datasets\titanic.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
# Applying Imputation - filling the null values in columns having null values.

si_age = SimpleImputer() #default imputation using mean value
si_emb = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_emb = si_emb.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_emb = si_emb.transform(X_test[['Embarked']])
```

```python
# one hot encoding on sex and embarked columns 

ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_emb = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_emb = ohe_emb.fit_transform(X_train_emb)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_emb = ohe_emb.transform(X_test_emb)
```

```python
# after the above code execution we have successfully brought the following changes
# 1. Empty values filled in age and embarked
# 2. Sex and embarked columns are now One Hot Encoded
# Hence, sex, age, and embarked are now transformed
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_emb),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_emb),axis=1)
```

```python
X_train_transformed.shape
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
